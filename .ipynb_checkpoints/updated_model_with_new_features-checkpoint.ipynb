```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('data_2020_06_01.xlsx', sheet_name='Sheet1', skiprows=[0])
df.head()
df.columns
features= ['Epidemiology', 'Fever', 'URTS', 'LRTS', 'GIS',
          'Flu like symptoms', 'Extreme fatigue', 'CXR', 'Leukocytes count',
         ' diagnosis','age', 'gender', 'resedens', 'CV', 'DM', 
           'HTN', 'LUNG']

target = 'PCR'
other_predictors = ['score']

x = df[features]
y=df[target]
y.replace('Positive',1, inplace=True)
y.replace('Negative',0, inplace=True)
print(y.mean(), len(y))
x.columns = ['Epidemiology', 'Fever', 'URTS', 'LRTS', 'GIS',
          'Flu like symptoms', 'Extreme fatigue', 'CXR', 'Leukocytes count',
         'Other Likely diagnosis','age', 'gender', 'resedens', 'CV', 'DM', 
           'HTN', 'LUNG']

x.loc[x['resedens']==2,'resedens']=1
df.shape
```

```python
# for col in x.columns:
#     x[col].hist()
#     plt.suptitle(col)
#     plt.show()
    
binary = ['Epidemiology', 'Fever', 'URTS', 'LRTS', 'GIS',
          'Flu like symptoms', 'Extreme fatigue', 'Other Likely diagnosis','gender',
          'CV', 'DM', 'HTN', 'LUNG']

categorical = ['CXR', 'Leukocytes count','resedens']

continuous = ['age']

```

```python
for col in categorical:

    print(x.groupby(col).size())
```

```python

for col in binary:
    x[col].fillna(0,inplace=True)


for col in  categorical:   
    cat_dummies = pd.get_dummies(x[col])
    x = x.merge(cat_dummies, how='inner', left_index=True , right_index=True)
    if col in x.columns:
        del x[col]


y.fillna(0,inplace=True)

np.round(x.describe().ix['mean'],2).sort_values().plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
     x, y, test_size=0.25, random_state=1)

from sklearn.model_selection import cross_validate
# # Explore Random Forest
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=300, 
#                              max_depth=15,
#                              class_weight = 'balanced',
                             random_state=0,
                             warm_start=True,
                             bootstrap=True,
                            min_impurity_decrease = 0.001)
cv_results_rf = cross_validate(clf, X_train, y_train, cv=5, scoring='recall')

from sklearn.linear_model import LogisticRegression
clf_lr = LogisticRegression()
cv_results_lr = cross_validate(clf_lr, X_train, y_train, cv=3,scoring='recall')

from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
cv_results_nb = cross_validate(clf_lr, X_train, y_train, cv=3,scoring='accuracy')

print('Cross Validatio Results for - Random Forest:' , cv_results_rf['test_score'].mean())
print('Cross Validatio Results for - Linear REgression:' , cv_results_lr['test_score'].mean())
print('Cross Validatio Results for - Linear REgression:' , cv_results_nb['test_score'].mean())



final_clf_rf = clf.fit(X_train,y_train)
final_clf_lr = clf_lr.fit(X_train,y_train)
```

```python
def plot_roc(clf):
    from sklearn.metrics import roc_auc_score
    from sklearn.metrics import roc_curve
    logit_roc_auc = roc_auc_score(y_test, clf.predict(X_test))
    fpr, tpr, thresholds = roc_curve(y_test, clf.predict_proba(X_test)[:,1])
    plt.figure()
    plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    plt.savefig('Log_ROC')
    plt.show()
    

def calc_errors_Accuracy(df_pred,y_test):
    errors = abs(df_pred - y_test)
    print("Accuracy:",round(1-np.mean(errors),2))
    
def plot_precision_recall_curve(df_actual,precision,recall):
    # plot the precision-recall curves
    no_skill = len(df_actual[df_actual == 1]) / len(df_actual)
    plt.plot([0, 1], [no_skill, no_skill], linestyle='--', label='No Skill')
    plt.plot(recall, precision, marker='.', label='NN')
    # axis labels
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    # show the legend
    plt.legend()
    # show the plot
    plt.show()
```

```python
# Get numerical feature importances
importances = list(clf.feature_importances_)
# List of tuples with variable and importance
feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(x.columns, importances)]
# Sort the feature importances by most important first
feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*pair)) for pair in feature_importances];


from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc

print()
print()
print()
print('Results from simple Random Forest')
df_pred = final_clf_rf.predict_proba(X_test)
df_actual = y_test
precision, recall, thresholds = precision_recall_curve(df_actual, df_pred[:,1])
df_pred = final_clf_rf.predict(X_test)
f1 = f1_score(df_actual, df_pred)
lr_f1, lr_auc = f1_score(df_actual, df_pred), auc(recall, precision)

from sklearn.metrics import recall_score,precision_score,confusion_matrix
print('RF: f1=%.3f auc=%.3f' % (lr_f1, lr_auc))
print('RF: Recall Score=%.3f ' % (recall_score(df_actual, df_pred)))
print('RF: Precision Score =%.3f ' % (precision_score(df_actual, df_pred)))
calc_errors_Accuracy(df_pred,y_test)
print('RF: confusion_matrix =' )
print(confusion_matrix(df_actual, df_pred))
plot_roc(final_clf_rf)
plot_precision_recall_curve(df_actual,precision,recall)


###########################################################
print()
print()
print()
print('Results from simple Logistic Regression')
df_pred = clf_lr.predict_proba(X_test)
df_actual = y_test
precision, recall, thresholds = precision_recall_curve(df_actual, df_pred[:,1])
df_pred = clf_lr.predict(X_test)
f1 = f1_score(df_actual, df_pred)
lr_f1, lr_auc = f1_score(df_actual, df_pred), auc(recall, precision)

from sklearn.metrics import recall_score,precision_score,confusion_matrix
print('LR: f1=%.3f auc=%.3f' % (lr_f1, lr_auc))
print('LR: Recall Score=%.3f ' % (recall_score(df_actual, df_pred)))
print('LR: Precision Score =%.3f ' % (precision_score(df_actual, df_pred)))
calc_errors_Accuracy(df_pred,y_test)
print('LR: confusion_matrix =' )
print(confusion_matrix(df_actual, df_pred))
plot_roc(clf_lr)
plot_precision_recall_curve(df_actual,precision,recall)
```
